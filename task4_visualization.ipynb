```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# Load data
df = pd.read_csv('/content/trends_analysed.csv')

# Create outputs folder
os.makedirs('outputs', exist_ok=True)

# CHART 1: Top 10 Stories
top_10 = df.nlargest(10, 'score')
top_10['title'] = top_10['title'].str[:50]  # Shorten titles

plt.figure(figsize=(10, 6))
plt.barh(top_10['title'], top_10['score'])
plt.xlabel('Score')
plt.ylabel('Title')
plt.title('Top 10 Stories by Score')
plt.savefig('outputs/chart1_top_stories.png')
plt.close()
```

```python
# ===== CHART 2: Stories per Category =====
category_counts = df['category'].value_counts()

plt.figure(figsize=(10, 6))
plt.bar(category_counts.index, category_counts.values)
plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Stories per Category')
plt.xticks(rotation=45)
plt.savefig('outputs/chart2_categories.png')
plt.close()

# ===== CHART 3: Score vs Comments =====
plt.figure(figsize=(10, 6))
plt.scatter(df[df['is_popular']==False] ['score'],
            df[df['is_popular']==False] ['num_comments'],
            label='Non-Popular', alpha=0.6)
plt.scatter(df[df['is_popular']==True] ['score'],
            df[df['is_popular']==True] ['num_comments'],
            label='Popular', alpha=0.6)
plt.xlabel('Score')
plt.ylabel('Comments')
plt.title('Score vs Comments')
plt.legend()
plt.savefig('outputs/chart3_scatter.png')
plt.close()

# ===== DASHBOARD =====
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Chart 1
axes[0].barh(top_10['title'], top_10['score'])
axes[0].set_title('Top 10 Stories by Score')

# Chart 2
axes[1].bar(category_counts.index, category_counts.values)
axes[1].set_title('Stories per Category')
axes[1].tick_params(axis='x', rotation=45)

# Chart 3
axes[2].scatter(df[df['is_popular']==False] ['score'],
                df[df['is_popular']==False] ['num_comments'],
                label='Non-Popular', alpha=0.6)
axes[2].scatter(df[df['is_popular']==True] ['score'],
                df[df['is_popular']==True] ['num_comments'],
                label='Popular', alpha=0.6)
axes[2].set_title('Score vs Comments')
axes[2].legend()

fig.suptitle('TrendPulse Dashboard', fontsize=16)
plt.savefig('outputs/dashboard.png')
plt.close()

print("Done! All charts saved.")
```
